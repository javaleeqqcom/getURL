```python
import re,os
from test import 中文小说处理类
import chinese2digits as c2d
ctool=中文小说处理类()
source_file="./fanren/all.txt"
output_folder="./fanren/split/"
def write_lines(path:str,lines):
    # try:
        with open(path,"w",encoding="gbk") as ofile:
            ofile.writelines(lines)
        return True
    # except:return False
```

```python
with open(source_file,"r",encoding="gbk") as rfile:
    content=rfile.readlines()

write_fail_t=0
last_i,book,chapter = 0,0,0
def write_chapter(i:int,title:str):
    global last_i,book,chapter,content,write_fail_t
    file_name="b{}_c{} {}.txt".format(book,chapter,title)
    if not write_lines(output_folder+file_name,content[last_i:i]):
        print("write_fail,book={},chapter={}".format(book,chapter))
        write_fail_t+=1
    else:
        print("{}\t{}\t{}".format(book,chapter,title))
    last_i=i

title=""
for i,line in enumerate(content):
    if '\n'==line[-1]:line=line[:-1]
    m=re.match(ctool.第几卷,line)
    if m!=None:
        write_chapter(i,title)
        book += 1
        chapter=0
        title=line
    else:
        m=re.match(ctool.第几章,line)
        if m!=None:
            write_chapter(i,title)
            chapter += 1
            title=line
write_chapter(len(content),title)
print("write_fail_times=",write_fail_t)
```
